## Import and clean dataset

Next, the dataset must be imported as a dataframe.

```python
stroke_dataset_df = pd.read_csv("data_Files/RawData/healthcare-dataset-stroke-data.csv")
stroke_dataset_df
```

---

Next, the information of the dataset is required to determine the columns and their data types.

```python
stroke_dataset_df.info()
```

Next, it must be determined if any columns have any null values.

```python
stroke_dataset_df.isnull().sum()
```

The only column that contained any null values was the "BMI" column. These rows can be dropped as BMI can not be imputed based off the other columns in the dataset.

```python
stroke_dataset_df= stroke_dataset_df.dropna()
stroke_dataset_df.isnull().sum()
```

Now there are no null values in the data, the next thing to do is to remove the "id" column. This column will not be relevant to any of the visualisation or analysis and so it is better to remove this.

```python
stroke_dataset_df = stroke_dataset_df.drop(labels=["id"], axis = 1)
stroke_dataset_df.info()
```

Next, it is important to make sure the data has the correct types. "Age" is currently a float, however, it should be an integer for.

```python
stroke_dataset_df = stroke_dataset_df.astype({"age":"int"})
stroke_dataset_df.info()
```

```python
stroke_dataset_df
```

From the source where the dataset is derived, it was mentioned that the "Gender" column has a third value of "Other" and the "Smoking status" column has an extra value of "unknown". The value of "other" has to be removed as it is too vague an answer and so incorrect correlations could be derived from generalising "other". "Unknown" must be removed as it will provide no value to predicting stroke likelihood. By knowing if a patient was a smoker or not can help determine the impact of smoking on stroke occurrence probability.

```python
stroke_dataset_df = stroke_dataset_df[stroke_dataset_df["gender"] != "Other"]
stroke_dataset_df["gender"].unique()
```

By running the "unique()" method we can see that the only remaining unique values for the "gender" column

```python
stroke_dataset_df = stroke_dataset_df[stroke_dataset_df["smoking_status"] != "Unknown"]
stroke_dataset_df["smoking_status"].unique()
```

The same method has been implemented for the "Smoking status" column to remove the "Unknown" values.

```python
stroke_dataset_df.info()
```

```python
stroke_dataset_df.head()
```

```python
stroke_dataset_df.tail()
```

```python
stroke_dataset_df.duplicated().sum()
```

A final assessment of the data is completed with snippets of the top and bottom of the datasets using the "head()" and "tail()" methods and a final check for any duplicated data.

# Conclusion

The initial data was not too complex to clean and so the rationale and process chosen was as follows:

* After the first null count analysis it was determined that only the "BMI" column had any null values. These rows had to be removed as "BMI" can not be predicted based off the other variables present in the dataset.
* Once the null values were removed, the dataset was analysed again on the datatypes of each variable. "Age" was a float and this was the only datatype that needed changing to an integer as "Age" is not a continuous variable.
* The next change implemented was the removal of the "id" column as this provides no value for modelling or analysis and will only cause issues later on down the line.
* It was then determined that both the "Gender" and "Smoking status" columns contained data values ("other" and "unknown" respectively) that were too vague and did not provide specific information on the impact on stroke probability. As a result, the rows containing these values were removed.
* Once all these changes were done, the dataset had one final analysis (head, tail and duplicates) and was exported as a clean data set ready for visualising the data in the next step
