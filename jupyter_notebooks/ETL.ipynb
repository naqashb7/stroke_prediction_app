## Import and clean dataset

Next, the dataset must be imported as a dataframe.

```python
stroke_dataset_df = pd.read_csv("data_Files/RawData/healthcare-dataset-stroke-data.csv")
stroke_dataset_df
```

---

Next, the information of the dataset is required to determine the columns and their data types.

```python
stroke_dataset_df.info()
```

Next, it must be determined if any columns have any null values.

```python
stroke_dataset_df.isnull().sum()
```

The only column that contained any null values was the "BMI" column. These rows can be dropped as BMI can not be imputed based off the other columns in the dataset.

```python
stroke_dataset_df= stroke_dataset_df.dropna()
stroke_dataset_df.isnull().sum()
```

Now there are no null values in the data, the only remaining cleaning is to remove the "id" column. This column will not be relevant to any of the visualisation or analysis and so it is better to remove this.

```python
stroke_dataset_df = stroke_dataset_df.drop(labels=["id"], axis = 1)
stroke_dataset_df.info()
```

Finally, it is important to make sure the data has the correct types. "Age" is currently a float, however, it should be an integer for.

```python
stroke_dataset_df = stroke_dataset_df.astype({"age":"int"})
stroke_dataset_df.info()
```

---

## Export Clean Data

```python
stroke_dataset_df.to_csv("data_files/CleanData/stroke_data_clean.csv", index=True)
```

The dataset has been cleaned and exported to the "Clean Data" folder ready for visualisation.

# Conclusion

The initial data was not too complex to clean and so the rationale and process chosen was as follows:

* After the first null count analysis it was determined that only the "BMI" column had any null values. These rows had to be removed as "BMI" can not be predicted based off the other variables present in the dataset.
* Once the null values were removed, the dataset was analysed again on the datatypes of each variable. "Age" was a float and this was the only datatype that needed changing to an integer as "Age" is not a continuous variable.
* The final change implemented was the removal of the "id" column as this provides no value for modelling or analysis and will only cause issues later on down the line.
* Once all these changes were done, the dataset had one final analysis and was exported as a clean data set ready for visualising the data in the next step
